```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
Wines=pd.read_csv('wine.csv')
```

```python
Wines
```

```python
Wines['Type'].value_counts()
```

```python
Wines1=Wines.iloc[:,1:]
Wines1
```

```python
wine_array=Wines1.values
wine_array
```

```python
data_norm=scale(wine_array)
```

```python
data_norm
```

```python
pca=PCA(n_components=13)

wine_pca=pca.fit_transform(data_norm)
wine_pca
```

```python
pca.components_
```

```python
var=pca.explained_variance_ratio_
var
```

```python
var1=np.cumsum(np.round(var,4)*100)
```

```python
var1
```

```python
plt.plot(var1,color='red')
```

```python
final_df=pd.concat([Wines['Type'],pd.DataFrame(wine_pca[:,0:3],columns=['PC1','PC2','PC3'])],axis=1)
final_df
```

```python
fig=plt.figure(figsize=(18,14))
sns.scatterplot(data=final_df)
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(10,8))
dendrogram=sch.dendrogram(sch.linkage(final_df,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=4,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(final_df),columns=['clusterid'])
y['clusterid'].value_counts()
```

```python
Wines2=Wines.copy()
Wines2['clusterid']=hclusters.labels_
Wines2
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range (1,9):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(final_df)
    wcss.append(kmeans.inertia_)
```

```python
 plt.plot(range(1,9),wcss)
plt.title('Elbow Graph')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
clust3=KMeans(3,random_state=32).fit(final_df)
clust3
```

```python
clust3.labels_
```

```python
Wines3=Wines.copy()
Wines3['clustid3']=clust3.labels_
```
